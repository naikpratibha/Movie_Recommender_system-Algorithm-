```python
#The dataset used in this demonstration is the TMDB 5000 Movie Dataset(Metadata on ~5000 movies on TMDB.
films=pd.read_csv('tmdb_5000_movies.csv')
credit=pd.read_csv('tmdb_5000_credits.csv')
```

```python
#By this we merge the two datas by title column
films = films.merge(credit,on='title')
```

```python
#We removed all the columns which are not required for the data analysis 
#genres
#id
#keywords
#title
#overview
#cast
#crew

films=films[['id','title','genres','keywords','overview','cast','crew']]
```

```python
#Here Keywords column is an array of dictionary and we derive the name of the keywords from it.
#'[{"id": 28, "name": "Action"}, {"id": 12, "name": "Adventure"}, {"id": 14, "name": "Fantasy"}, {"id": 878, "name": "Science Fiction"}]'
#['Action','Adventure','Fantasy','Science Fiction']
```

```python
import ast
```

```python
#function to fetch names from the given column
def convert(obj):
    L = []
    for i in ast.literal_eval(obj):
        L.append(i['name'])
    return L
```

```python
films['genres'] = films['genres'].apply(convert)
```

```python
films.head(2)
```

```python
films['keywords'] = films['keywords'].apply(convert)
```

```python
films.head(2)
```

```python
#function to fetch top 3 actor/actress name from the cast.​
def convert3(text):
    L = []
    counter = 0
    for i in ast.literal_eval(text):
        if counter < 3:
            L.append(i['name'])
        counter+=1
    return L 
```

```python
films['cast'] = films['cast'].apply(convert)
films.head(2)
```

```python
films['cast'] = films[ 'cast'].apply(lambda x:x[0:3])
```

```python
#Function to fetch director name from the crew column.​
def bring_director(text):
    L = []
    for i in ast.literal_eval(text):
        if i['job'] == 'Director':
            L.append(i['name'])
    return L 
```

```python
films['crew'] = films['crew'].apply(bring_director)
```

```python
films.head(2)
```

```python
films['overview'][0]
```

```python
#We split the overview strings into list form
films['overview'].apply(lambda x:x.split())
```

```python
films['overview'] = films['overview'].apply(lambda x:x.split())
```

```python
films.head(2)
```

```python
#Removing spaces to avoid any multiple counts 
films['genres'] = films['genres'].apply(lambda x:[i.replace(" ","") for i in x])
films['keywords'] =films['keywords'].apply(lambda x:[i.replace(" ","") for i in x])
films['cast'] =films['cast'].apply(lambda x:[i.replace(" ","") for i in x])
films['crew'] = films['crew'].apply(lambda x:[i.replace(" ","") for i in x])
```

```python
films.head(2)
```

```python
films['labels'] = films['overview'] + films['genres'] + films['keywords'] + films['cast'] + films['crew']
```

```python
films.head(2)
```

```python
#We create a new table With columns id,titlw and labels
new_table= films[['id', 'title', 'labels']]
```

```python
new_table['labels'] = new_table['labels'].apply(lambda x:" ".join(x))
```

```python
 new_table.head(2)
```

```python
import nltk
```

```python
from nltk.stem.porter import PorterStemmer
pts = PorterStemmer()
```

```python
def stem(text):
    p = []
    
    for i in text.split():
       p.append(pts.stem(i))
    return " ".join(p)
```

```python
new_table['labels'] = new_table['labels'].apply(stem)
```

```python
#VECTORISATION OF LABELS
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features=5000, stop_words='english')
```

```python
vect = cv.fit_transform(new_table['labels']).toarray()
```

```python
vect
```

```python
vect[0]
```

```python
cv.get_feature_names()
```

```python
#We use cosine_similarity function to calculate daistance of one movie with other movies and to compare them.
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similar = cosine_similarity(vect)
```

```python
sorted(similar[0], reverse=True)
```

```python
similar[1]
```

```python
new_table['labels']
```

```python
#A function which will return the name of the 5 films with maximum similarity with the given movie.
def Suggest(film):
    films_ind =  new_table[new_table['title'] ==film].index[0]
    distances = similar[films_ind]
    films_list = sorted(list(enumerate(distances)), reverse=True,key=lambda x:x[1])[1:6]
    
    for i in films_list:
        print(new_table.iloc[i[0]].title)
```

```python
Suggest('The Hobbit: The Desolation of Smaug')
```
